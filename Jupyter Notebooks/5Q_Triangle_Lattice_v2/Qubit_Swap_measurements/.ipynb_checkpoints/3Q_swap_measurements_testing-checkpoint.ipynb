```python
def acquire_data(filepath):
    with h5py.File(filepath, "r") as f:
        
        time_units = 2.32515 / 16
        
        times = f['wait_times'][()]
        exp_data = f['Exp_values'][0,:]
        
    times *= time_units
    
    return times, exp_data
```

```python
def generate_swaps_filename(year, month, day, hour, minute, second, local=False):
    date_code = f'{year}_{month}_{day}'
    time_code = f'{hour}_{minute}_{second}'
    if local:
        return r'C:\Users\mattm\OneDrive\Desktop\Research\Measurements\5QV2_Triangle_Lattice\QubitOscillationsMUX\QubitOscillationsMUX_{}\QubitOscillationsMUX_{}_{}_data.h5'.format(date_code, date_code, time_code)
    else:
        return r'V:\QSimMeasurements\Measurements\5QV2_Triangle_Lattice\QubitOscillationsMUX\QubitOscillationsMUX_{}\QubitOscillationsMUX_{}_{}_data.h5'.format(date_code, date_code, time_code)
    
```

```python
class QubitSwapTraceMeasurement:
    
    def __init__(self, readout_qubit, filename, **kwargs):
        '''
        Measurement class for 3Q swaps
        :param readout_qubit: qubit whose population will be readout
        :param filename: complete filename of swap data
        :param kwargs: used to add meta data for measurement object, such as what voltage/frequency the qubit/coupler was at
        
        '''
        
        self.readout_qubit = readout_qubit
        self.filename = filename
        
        self.times = None
        self.populations = None
        
        self.times = None
        self.populations = None
        
        self.meta_data_dict = kwargs
        
    def get_times(self):
        if self.times is None:
            self.acquire_data()
        return self.times
    
    def get_populations(self):
        if self.populations is None:
            self.acquire_data()
        return self.populations
        
    def acquire_data(self):
        self.times, self.populations = acquire_data(self.filename)
        
    def plot_trace(self):
        
        times = self.get_times()
        populations = self.get_populations()
        
        plt.plot(times, populations, linestyle='', marker='o', ms=4)
        
        plt.xlabel('time (ns)')
        plt.ylabel('Population')
        
        plt.title(f'{self.readout_qubit} swap trace')
        plt.show()
```

```python
def generate_Hamiltonian(g_12, g_23, g_13, U):
    
    H = omega_1 * a1.dag()*a1 + omega_2 * a2.dag()*a2 + omega_3 * a3.dag()*a3 
    H += g_12*(a1*a2.dag() + a2*a1.dag()) + g_23*(a3*a2.dag() + a2*a3.dag()) + g_13*(a1*a3.dag() + a3*a1.dag())
    
    H += U * (a1.dag()*a1*(a1.dag()*a1 - 1) + a2.dag()*a2*(a2.dag()*a2 - 1) + a3.dag()*a3*(a3.dag()*a3 - 1)) 
    
    return H
```

```python
def cost_function(x, data, times, psi0, e_ops=None, c_ops=None):
    
    g_12, g_23, g_13, gamma_1, gamma_phi = x
    
    H = generate_Hamiltonian(g_12, g_23, g_13, U)
    
    if e_ops is None:
        e_ops = [a1.dag()*a1, a2.dag()*a2, a3.dag()*a3]
    
    if c_ops is None:
        c_ops = [np.sqrt(gamma_1)*a1, np.sqrt(gamma_1)*a2, np.sqrt(gamma_1)*a2,
                 np.sqrt(gamma_phi)*a1.dag()*a1, np.sqrt(gamma_phi)*a2.dag()*a2, np.sqrt(gamma_phi)*a3.dag()*a3]
    
    result = qt.mesolve(H, psi0, times/1e3, e_ops=e_ops, c_ops=c_ops)
    
    cost = 0
    for i in range(len(result.expect)):
        cost += np.sum(np.power(result.expect[i] - data[i], 2))
    
    return cost
```

```python
def cost_function_single_qubit_decoherence(x, g_12, g_23, g_13, data, times, psi0, e_ops=None, c_ops=None):
    
    gamma_1, gamma_phi = x
    
    H = generate_Hamiltonian(g_12, g_23, g_13, U)
    
    if e_ops is None:
        e_ops = [a1.dag()*a1, a2.dag()*a2, a3.dag()*a3]
    
    if c_ops is None:
        c_ops = [np.sqrt(gamma_1)*a1, np.sqrt(gamma_1)*a2, np.sqrt(gamma_1)*a2,
                 np.sqrt(gamma_phi)*a1.dag()*a1, np.sqrt(gamma_phi)*a2.dag()*a2, np.sqrt(gamma_phi)*a3.dag()*a3]
    
    result = qt.mesolve(H, psi0, times/1e3, e_ops=e_ops, c_ops=c_ops)
    
    return np.sum(np.power(result.expect[0] - data[i], 2))
    
```

```python
def run_simulation(g_12, g_23, g_13, gamma_1, gamma_phi, U, num_levels, times, initial_state='plus', e_ops=None, c_ops=None):
    
    a = qt.destroy(num_levels)

    a1 = qt.tensor([a, qt.qeye(num_levels), qt.qeye(num_levels)])
    a2 = qt.tensor([qt.qeye(num_levels), a, qt.qeye(num_levels)])
    a3 = qt.tensor([qt.qeye(num_levels), qt.qeye(num_levels), a])
    
    if initial_state == 'plus':
        psi0 = 1/np.sqrt(2)*(qt.basis([num_levels, num_levels, num_levels], [1, 0, 0]) + qt.basis([num_levels, num_levels, num_levels], [0, 0, 1]))
    elif initial_state == 'minus':
        psi0 = 1/np.sqrt(2)*(qt.basis([num_levels, num_levels, num_levels], [1, 0, 0]) - qt.basis([num_levels, num_levels, num_levels], [0, 0, 1]))
        
    H = generate_Hamiltonian(g_12, g_23, g_13, U)
    
    
    if e_ops is None:
        e_ops = [a1.dag()*a1, a2.dag()*a2, a3.dag()*a3]
    
    if c_ops is None:
        c_ops = [np.sqrt(gamma_1)*a1, np.sqrt(gamma_1)*a2, np.sqrt(gamma_1)*a2,
                 np.sqrt(gamma_phi)*a1.dag()*a1, np.sqrt(gamma_phi)*a2.dag()*a2, np.sqrt(gamma_phi)*a3.dag()*a3]
    
    result = qt.mesolve(H, psi0, times/1e3, e_ops=e_ops, c_ops=c_ops)
    
    populations = np.zeros((len(e_ops), len(times)))
    for i in range(len(e_ops)):
        populations[i,:] = result.expect[i]
        
    return populations
    
```

```python
num_levels = 3

U = 180 * 2 * np.pi # MHz

g_12_positive = 7.49 * 2 * np.pi # 2pi MHz
g_23_positive = 7.41 * 2 * np.pi # 2pi MHz

g_12_positive = 7.49 * 2 * np.pi * 0.931 # 2pi MHz
g_23_positive = g_12

g_13_positive = 9.48 * 2 * np.pi # 2pi MHz
# g_13 = 5.93510072 * 2 * np.pi # 2pi MHz

gamma_1_positive = 0.07185956 # MHz

gamma_phi_positive = 0.14038149 # MHz

```
