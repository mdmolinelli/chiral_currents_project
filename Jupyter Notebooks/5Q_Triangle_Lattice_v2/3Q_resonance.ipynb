```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
couplings = np.linspace(-30, 30, 101) # MHz

g = 10 # MHz

eigenvalue_data = np.zeros((len(couplings), 3))

for i in range(len(couplings)):
    
    g_eff = couplings[i]
    
    H = np.array([[0, g, g_eff],
                  [g, 0, g],
                  [g_eff, g, 0]])
    
    
    eigenvalues, eigenvectors = np.linalg.eig(H)
    eigenvalue_data[i, :] = np.sort(eigenvalues)
    
```

```python
for i in range(eigenvalue_data.shape[1]):
    plt.plot(couplings, eigenvalue_data[:,i])
    
plt.xlabel('Leg coupling (MHz)')
plt.ylabel('Frequencies (MHz)')

plt.axvline(g, linestyle=':', color='black', label='rung coupling (MHz)')
plt.axvline(-g, linestyle=':', color='black')

plt.legend()
plt.grid()

plt.show()
```
